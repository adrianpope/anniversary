```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from importlib import reload
```

```python
from matplotlib import rc
rc('font', size=22)
rc('text', usetex=True)
rc('text.latex', preamble=r'\usepackage{wasysym}\usepackage{fontawesome}')
```

```python
def orbit(t, r, p):
    x = -1.*r*np.sin(2*np.pi*t/p)
    y = r*np.cos(2*np.pi*t/p)
    return x,y
```

```python
r_earth = 1.0
r_moon = 0.12
fmfudge = 0.999
fss = 120
fsm = 36
fse = 24
cini = 0.9
cfin = 0.0
#
p_earth = 364
p_moon = 26
#
to = np.linspace(0,p_earth,p_earth+1)
xoe,yoe = orbit(to, r_earth, p_earth)
xom,yom = orbit(to, r_moon, p_moon)
#
tf = np.arange(13)/13.*p_earth
xfe,yfe = orbit(tf, r_earth, p_earth)
xfm,yfm = orbit(tf, r_moon, p_moon)
#
plt.figure(figsize=(8,8))
plt.plot(xoe, yoe, '--k')
plt.plot(xoe + xom, yoe + yom, ':k')
#
#plt.text(0,0,r'\sun',ha='center',va='center',fontsize=fss)
plt.text(0,0,r'\faSunO',ha='center',va='center',fontsize=fss)
for i in range(len(tf)):
    x = fmfudge*(xfe[i] + xfm[i])
    y = fmfudge*(yfe[i] + yfm[i])
    rot = np.arctan2(y,x)*180./np.pi
    gray = '%.3f'%(cini + (cfin-cini)*i/(1.0*len(tf)-1))
    plt.text(x,y,r'\RIGHTcircle',ha='center',va='center',rotation=rot,fontsize=fsm,c=gray)
    #plt.text(xfe[i],yfe[i],r'$\oplus$',ha='center',va='center',fontsize=fse)
#
plt.axis('off')
plt.tight_layout()
plt.savefig('1year.pdf')
plt.show()
```
